```python
import requests
from bs4 import BeautifulSoup
import json
import time

# Danh s√°ch c√°c m√£ ng√†nh quan tr·ªçng c·ªßa EAIT
EAIT_PREFIXES = ["CSSE", "INFS", "ENGG", "COMP", "DECO", "CYBR", "DATA", "ELEC", "MECH", "CIVL"]

def fetch_course_codes_via_search(prefix):
    # D√πng trang SEARCH thay v√¨ Browse
    url = f"https://my.uq.edu.au/programs-courses/search.html?keywords={prefix}&searchType=course"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    codes = set()
    
    print(f"   -> ƒêang t√¨m ki·∫øm t·ª´ kh√≥a: {prefix} t·∫°i {url}")
    try:
        response = requests.get(url, headers=headers)
        
        if response.status_code != 200:
            print(f"      ‚ö†Ô∏è L·ªói k·∫øt n·ªëi: {response.status_code}")
            return []

        soup = BeautifulSoup(response.text, 'html.parser')
        
        # T√¨m c√°c link trong b·∫£ng k·∫øt qu·∫£
        # M·∫πo: C√°c link m√¥n h·ªçc lu√¥n ch·ª©a "course_code="
        links = soup.find_all('a', href=True)
        
        for link in links:
            href = link['href']
            if "course_code=" in href:
                # URL th∆∞·ªùng l√†: /programs-courses/course.html?course_code=CSSE1001&...
                # T√°ch l·∫•y m√£ m√¥n
                try:
                    # L·∫•y ph·∫ßn sau d·∫•u = ƒë·∫ßu ti√™n v√† tr∆∞·ªõc d·∫•u & (n·∫øu c√≥)
                    part = href.split("course_code=")[1]
                    code = part.split("&")[0]
                    
                    # Ki·ªÉm tra l·∫°i: M√£ ph·∫£i b·∫Øt ƒë·∫ßu b·∫±ng Prefix v√† d√†i 8 k√Ω t·ª±
                    if code.startswith(prefix) and len(code) == 8:
                        codes.add(code)
                except:
                    continue
                    
    except Exception as e:
        print(f"L·ªói khi qu√©t {prefix}: {e}")
        
    return list(codes)

# --- CH·∫†Y L·∫†I B∆Ø·ªöC 1 ---
all_codes = []
print("üöÄ B·∫ÆT ƒê·∫¶U B∆Ø·ªöC 1 (FIXED): QU√âT M√É M√îN...")

for prefix in EAIT_PREFIXES:
    found = fetch_course_codes_via_search(prefix)
    print(f"      ‚úÖ T√¨m th·∫•y {len(found)} m√¥n {prefix}.")
    all_codes.extend(found)
    time.sleep(1) # Ngh·ªâ 1 gi√¢y

# L·ªçc tr√πng v√† s·∫Øp x·∫øp
all_codes = sorted(list(set(all_codes)))

# L∆∞u file
with open('data/eait_codes_only.json', 'w', encoding='utf-8') as f:
    json.dump(all_codes, f, indent=4)

print(f"\nüéâ T·ªîNG K·∫æT: ƒê√£ t√¨m th·∫•y {len(all_codes)} m√£ m√¥n ƒë·ªôc nh·∫•t.")
print("File ƒë√£ l∆∞u: eait_codes_only.json")
```
